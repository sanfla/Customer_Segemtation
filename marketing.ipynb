```python
yr = 2024
data['umur'] = yr-data['Year_Birth']

data["Tinngal_dengan"]=data["Marital_Status"].replace({"Married":"Partner", 
                                                    "Together":"Partner", 
                                                    "Absurd":"Alone", 
                                                    "Widow":"Alone", 
                                                    "YOLO":"Alone", 
                                                    "Divorced":"Alone", 
                                                    "Single":"Alone",})

data["anak"]=data["Kidhome"]+data["Teenhome"]

data["pendidikan"]=data["Education"].replace({"Basic":"Undergraduate",
                                             "2n Cycle":"Undergraduate", 
                                             "Graduation":"Graduate", 
                                             "Master":"Postgraduate", 
                                             "PhD":"Postgraduate"})

data['Pengeluaran_2thn'] = data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']

data['kemakan_iklan'] = data['AcceptedCmp1'] + data['AcceptedCmp2'] + data['AcceptedCmp3'] + data['AcceptedCmp4'] + data['AcceptedCmp5']

data['tempat_beli'] = data['NumWebPurchases'] + data['NumCatalogPurchases'] + data['NumStorePurchases'] + data['NumDealsPurchases']
```

```python
plt.figure(figsize=(15, 40))  
for i, j in zip(numerical, range(1, len(numerical)+1)):
    plt.subplot(10, 2, j) 
    sns.kdeplot(data[i], fill=True)
plt.show()
```

```python
def hapus_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

data = hapus_outliers(data, numerical)
```

```python
data = data.copy()
```

```python
categorical = data.select_dtypes(include=['object']).columns 
```

```python
encoded_data = pd.get_dummies(data, columns=categorical, drop_first=True, dtype=int) 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(encoded_data)
```

```python
from sklearn.cluster import KMeans

def elbow(X, max_cls):
    wcss = []
    for k in range(1, max_cls+1):
        kmeans = KMeans(n_clusters=k, n_init=10, random_state=42).fit(X)
        wcss.append(kmeans.inertia_)

    plt.figure(figsize = (6, 4))
    plt.plot(range(1,max_cls+1), wcss, marker='o')
    plt.title('The Elbow Method')
    plt.xlabel('Number of clusters')
    plt.ylabel('WCSS')
    plt.show()

elbow(scaled_data, 10)
```

```python
from sklearn.metrics import silhouette_score

range = [2, 3, 4, 5, 6, 7, 8, 9, 10]

for num_clusters in range:
    kmeans = KMeans(num_clusters, n_init=10, random_state=101)
    kmeans.fit(scaled_data)
    
    cluster_labels = kmeans.labels_

    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
kmeans = KMeans(n_clusters=3, n_init=10, max_iter=50)

kmeans.fit(scaled_data)

clusters = kmeans.fit_predict(scaled_data)

transform_data = scaler.inverse_transform(scaled_data)
```

```python
pd.DataFrame(transform_data, columns=encoded_data.columns).head()
```

```python
plt.figure(figsize= (10,6))

plt.scatter(transform_data[clusters == 0, 0], transform_data[clusters == 0, 6], s = 25, c = 'blue')
plt.scatter(transform_data[clusters == 1, 0], transform_data[clusters == 1, 6], s = 25, c = 'red')
plt.scatter(transform_data[clusters == 2, 0], transform_data[clusters == 2, 6], s = 25, c = 'green')

plt.title('Clusters of customers')
plt.xlabel('Income')
plt.ylabel('Pengeluaran')

plt.show()
```
